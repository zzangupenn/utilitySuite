#### ConfigYAML

```python
import utilsuite
config = utilsuite.ConfigYAML()
# Set attributes for the configuration and direct save/load with yaml files.
config.name = 'saika' # Example attribute
config.age = 25 # Example attributes
config.skills = ['Python', 'YAML', 'Data Analysis'] # Example list attribute
config.preferences = {
    'theme': 'bright',
} # Example dictionary attribute
config.save('saika_config.yaml') # Save the configuration to a file
config.load('saika_config.yaml') # Load the configuration from a file
```

#### ListDict

```python
import utilsuite
rec = utilsuite.ListDict() 
# Easy data recoding and save/load
rec.init('data1', 'data2')
rec.data1 = [1, 2, 3]
rec.data2 = [4, 5, 6]
rec.save_onefile(filename='record', compress=True) # Save the record to one file
# rec.save('data1', 'data2') # Save the record to files with default filename
rec2 = utilsuite.ListDict()
rec2.load_onefile(filename='record') # Load the record from a file
print(rec2.data1, rec2.data2)
```

#### pltUtils

```python
# Example usage of the modified_sigmoid function and pltUtils for plotting
import utilsuite
import numpy as np

x = np.linspace(-2, 2, 100)
plt = utilsuite.pltUtils()
axs = plt.get_fig((2, 1))
axs[0].plot(x, utilsuite.modified_sigmoid(x, 5, 0.5))
axs[0].set_xlabel('x')
axs[0].set_ylabel('modified_sigmoid(x, 5, 0.5)')
axs[0].set_title('Modified Sigmoid Function')
axs[1].plot(x, utilsuite.modified_sigmoid(x, 20, -1))
axs[1].set_xlabel('x')
axs[1].set_ylabel('modified_sigmoid(x, 20, -1)')
plt.show()
```
