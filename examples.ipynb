### ConfigYAML
Directly set attributes for the configuration and save/load directly with yaml files.

```python
from utilitySuite import ConfigYAML
config = ConfigYAML()
# Directly set attributes for the configuration and save/load directly with yaml files.
config.name = 'saika' # Example attribute
config.age = 25 # Example attributes
config.skills = ['Python', 'YAML', 'Data Analysis'] # Example list attribute
config.preferences = {
    'theme': 'bright',
    'notifications': True,
    'language': 'English'
} # Example dictionary attribute
config.save('saika_config.yaml') # Save the configuration to a file
config.load('saika_config.yaml') # Load the configuration from a file
```

```python
# Example usage of the modified_sigmoid function and pltUtils for plotting
import utilitySuite as us
import numpy as np

x = np.linspace(-2, 2, 100)
plt = us.pltUtils()
axs = plt.get_fig((2, 1))
axs[0].plot(x, us.modified_sigmoid(x, 5, 0.5))
axs[0].set_xlabel('x')
axs[0].set_ylabel('modified_sigmoid(x, 5, 0.5)')
axs[0].set_title('Modified Sigmoid Function')
axs[1].plot(x, us.modified_sigmoid(x, 20, -1))
axs[1].set_xlabel('x')
axs[1].set_ylabel('modified_sigmoid(x, 5, 0.5)')
plt.show()
```
